```python
from skimage import data #For importing data

from skimage import color #For the color maps

from skimage.filters import sobel #For getting the outlines

from skimage import io #For importing the data in the folder

from skimage import exposure 

from skimage.transform import rotate #For rotating the images

from skimage.transform import rescale #for  Rescaling images

from skimage.transform import resize #For resizing images

from skimage import morphology

import matplotlib.pyplot as plt #For displaying the images
```

```python
img = data.coins()

plt.imshow(img)
```

```python
edge_sobel = sobel(img)
plt.imshow(edge_sobel)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(img, cmap=plt.cm.gray)
ax1.set
ax1.axis('off')
ax2.imshow(edge_sobel, cmap=plt.cm.gray)
ax2.set
ax2.axis('off')
```

```python
Whitebeard = io.imread("whitebeard.jpg")
```

```python
plt.imshow(Whitebeard)
```

```python
graybeard = color.rgb2gray(Whitebeard)
rgb = color.gray2rgb(graybeard) 
```

```python
image_eq = exposure.equalize_hist(rgb)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(graybeard, cmap=plt.cm.gray)
ax1.set
ax1.axis('off')
ax2.imshow(image_eq, cmap=plt.cm.gray)
ax2.set
ax2.axis('off')
```

```python
sidebeard = rotate(Whitebeard, -90)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(Whitebeard, cmap=plt.cm.gray)
ax1.set
ax1.axis('off')
ax2.imshow(sidebeard, cmap=plt.cm.gray)
ax2.set
ax2.axis('off')
```

```python
othersidebeard = rotate(Whitebeard, 90)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(Whitebeard, cmap=plt.cm.gray)
ax1.set
ax1.axis('off')
ax2.imshow(othersidebeard, cmap=plt.cm.gray)
ax2.set
ax2.axis('off')
```

```python
degreebeard = rotate(Whitebeard, 15)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(Whitebeard, cmap=plt.cm.gray)
ax1.set
ax1.axis('off')
ax2.imshow(degreebeard, cmap=plt.cm.gray)
ax2.set
ax2.axis('off')
```

```python
widebeard = rescale(Whitebeard, 1/4)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(Whitebeard)
ax1.set
ax1.axis('off')
ax2.imshow(widebeard)
ax2.set
ax2.axis('off')
```

```python
plt.imshow(widebeard)
```

```python
height = 900
width = 1000
sizebeard = resize(Whitebeard,(height, width), anti_aliasing=True)

plt.imshow(Whitebeard)
```

```python
plt.imshow(sizebeard)
```

```python
square = morphology.square(4)
plt.imshow(square)
```

```python
rectangle = morphology.rectangle(4, 2)
plt.imshow(rectangle)
```

```python
selem = morphology.rectangle(12,6)

erobeard = morphology.binary_erosion(graybeard, selem=selem)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
sharey=True)
ax1.imshow(graybeard, cmap=plt.cm.gray)
ax1.set
ax1.axis('off')
ax2.imshow(erobeard, cmap=plt.cm.gray)
ax2.set
ax2.axis('off')
```
